```python
import pandas as pd
from new_dict import name_reformat, column_reformat, date_reformat, fert_reformat, h_e_reformat, business_counts_reformat, business_register_reformat, disposable_income_reformat, jobs_density_reformat, population_reformat
```

```python
aps = pd.read_csv('ready_data/aps-districts-joined.csv')
aps = aps[aps.MEASURES_NAME == 'Variable']
del aps['MEASURES_NAME']
del aps['OBS_STATUS_NAME']
aps = aps.replace({"VARIABLE_NAME": name_reformat, "DATE_NAME": date_reformat})
aps.rename(columns=column_reformat, inplace=True)
aps.reset_index(drop=True, inplace=True)
aps.head(10)
```

```python
fertility = pd.read_csv('ready_data/fertility.csv')
del fertility['OBS_STATUS_NAME']
del fertility['MEASURES_NAME']
fertility = fertility.replace({"MEASURE_NAME": fert_reformat})
fertility.rename(columns=column_reformat, inplace=True)
fertility.head(10)
```

```python
h_e = pd.read_csv('ready_data/h-and-e-districts-joined.csv')
h_e = h_e[h_e.MEASURES_NAME == 'Value']
del h_e['MEASURES_NAME']
del h_e['OBS_STATUS_NAME']
del h_e['SEX_NAME']
h_e['indicator'] = h_e['PAY_NAME'] + ' ' + h_e['ITEM_NAME']
del h_e['PAY_NAME']
del h_e['ITEM_NAME']
h_e = h_e.replace({"indicator": h_e_reformat})
h_e.rename(columns=column_reformat, inplace=True)
h_e.head(10)
```

```python
business_counts = pd.read_csv('ready_data/business_counts.csv')
business_counts = business_counts[business_counts.LEGAL_STATUS_NAME == 'Total']
del business_counts['LEGAL_STATUS_NAME']
del business_counts['MEASURES_NAME']
del business_counts['OBS_STATUS_NAME']
business_counts['indicator'] = business_counts['INDUSTRY_NAME'] + ' ' + business_counts['EMPLOYMENT_SIZEBAND_NAME']
del business_counts['INDUSTRY_NAME']
del business_counts['EMPLOYMENT_SIZEBAND_NAME']
business_counts = business_counts.replace({"indicator": business_counts_reformat})
business_counts.rename(columns=column_reformat, inplace=True)
business_counts.head(10)
```

```python
business_register = pd.read_csv('ready_data/business_register_and_employment_survey.csv')
business_register = business_register[business_register['MEASURE_NAME'] == 'Count']
business_register['indicator'] = business_register['INDUSTRY_NAME'] + ' ' + business_register['EMPLOYMENT_STATUS_NAME']
del business_register['EMPLOYMENT_STATUS_NAME']
del business_register['MEASURES_NAME']
del business_register['MEASURE_NAME']
del business_register['INDUSTRY_NAME']
del business_register['OBS_STATUS_NAME']
business_register = business_register.replace({"indicator": business_register_reformat})
business_register.rename(columns=column_reformat, inplace=True)
business_register.head(10)
```

```python
disposable_income = pd.read_csv('ready_data/disposable_income.csv')
del disposable_income['OBS_STATUS_NAME']
del disposable_income['MEASURES_NAME']
del disposable_income['COMPONENT_OF_GDHI_NAME']
disposable_income.rename(columns=column_reformat, inplace=True)
disposable_income = disposable_income.replace({"indicator": disposable_income_reformat})
disposable_income.head(10)
```

```python
jobs_density = pd.read_csv('ready_data/jobs_density.csv')
del jobs_density['OBS_STATUS_NAME']
del jobs_density['MEASURES_NAME']
jobs_density.rename(columns=column_reformat, inplace=True)
jobs_density = jobs_density.replace({"indicator": jobs_density_reformat})
jobs_density.head(10)
```

```python
population = pd.read_csv('ready_data/population.csv')
population = population[population['GENDER_NAME'] == 'Total']
del population['OBS_STATUS_NAME']
del population['MEASURES_NAME']
del population['GENDER_NAME']
population.rename(columns=column_reformat, inplace=True)
population = population.replace({"indicator": population_reformat})
population.head(10)
```

```python
def swap_columns(df, col1, col2):
    col_list = list(df.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    df = df[col_list]
    return df

h_e = swap_columns(h_e, 'indicator', 'value')
business_counts = swap_columns(business_counts, 'indicator', 'value')
business_register = swap_columns(business_register, 'indicator', 'value')
```

```python
#  Confirm that there are no duplicate variable names
datasets = [aps, h_e, business_counts, business_register, fertility, population, jobs_density, disposable_income]
variable_names = []
for dataset in datasets:
    variable_names += list(dataset.indicator.unique())
len(variable_names) == len(set(variable_names)) 
print(variable_names)
```

```python
# Construct full dataset
final_dataset = pd.DataFrame
datasets = [aps, h_e, business_counts, business_register, fertility, population, jobs_density, disposable_income]
for dataset in datasets:    
    if final_dataset.empty:
        final_dataset = dataset
    else:
        final_dataset = pd.concat([final_dataset, dataset.iloc[1:]], ignore_index=True)     
```
